```python
#### Edge detection 

import matplotlib.pyplot as plt
import numpy as np

# Given data
lambdas = [0.0001, 0.000129, 0.000167, 0.000215, 0.000278, 0.000359, 0.000464, 0.000599, 0.000774, 0.001]
mae = [0.1823, 0.1781, 0.1606, 0.1688, 0.1652, 0.1947, 0.171, 0.1908, 0.1852, 0.1855]
parameter_sparsity = [58.0368, 65.8192, 82.7242, 83.8031, 80.9171, 80.8076, 71.0649, 80.117, 86.1968, 82.1553]

# Convert lambda to string for displaying purposes
lambda_str = [str(l) for l in lambdas]

# Create a new figure and a twin y-axis using subplot
plt.style.use('default')

# Recreate the plot with the specified style adjustments
fig, ax1 = plt.subplots()

# Plotting MAE with orange line
ax1.plot(lambda_str, mae, 'o-', color='red', label='MAE',zorder=10)

# Set x-axis label and rotate the tick labels for better visibility
ax1.set_xlabel('lambda')
ax1.set_xticklabels(lambda_str, rotation=45)

# Set y-axis label for MAE
ax1.set_ylabel(r'Mean Absolute Error, MAE ($\downarrow$)', color='black')

# Set y-axis tick params for MAE
ax1.tick_params(axis='y', labelcolor='black')

# Create a twin y-axis for Parameter sparsity
ax2 = ax1.twinx()

# Plotting Parameter sparsity with blue bars
ax2.bar(lambda_str, parameter_sparsity, alpha=0.3, color='blue', label='Parameter sparsity', zorder=1)

# Set y-axis label for Parameter sparsity
ax2.set_ylabel('Parameter sparsity', color='black')

# Set y-axis tick params for Parameter sparsity
ax2.tick_params(axis='y', labelcolor='black')

# Set title of the graph with black color
# plt.title('MAE and Parameter sparsity for different lambdas', color='black')
ax1.set_ylim(0.1, 0.22)  # For MAE
ax2.set_ylim(50, 95)     # For Parameter sparsity
# Show legend with black text
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Set all the text in the plot to black
for text in fig.findobj(plt.Text):
    text.set_color('black')


plt.text(1.2, 87, 'Edge detection (T4)', fontsize=12, ha='center', va='center')
# Tight layout to prevent clipping of tick labels
plt.tight_layout()
plt.savefig('outfiles/edge.png')
# Show the plot
plt.show()

```
